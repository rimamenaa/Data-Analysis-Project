```python
# drop index
ds_jobs_uncleaned.drop('index', inplace=True, axis=1)
```

```python
ds_jobs_uncleaned.head()
```

```python
ds_jobs_uncleaned['Salary Estimate'].unique()
```

```python
# cleaning the salary estimate column, removing what's inside the paranthesis
ds_jobs_uncleaned['Salary Estimate'] = ds_jobs_uncleaned['Salary Estimate'].apply(lambda x : x.split('(')[0])
```

```python
ds_jobs_uncleaned['Salary Estimate'].unique()
```

```python
# removing the 'K' and dollar sign
ds_jobs_uncleaned['Salary Estimate'] = ds_jobs_uncleaned['Salary Estimate'].apply(lambda x : x.replace('K','').replace('$',''))
```

```python
ds_jobs_uncleaned['Salary Estimate'].unique()
```

```python
# adding columns for the min and max salary and converting it to an integer
ds_jobs_uncleaned['min_salary'] = ds_jobs_uncleaned['Salary Estimate'].apply(lambda x : int(x.split('-')[0]))
ds_jobs_uncleaned['max_salary'] = ds_jobs_uncleaned['Salary Estimate'].apply(lambda x : int(x.split('-')[1]))
```

```python
ds_jobs_uncleaned.head()
```

```python
# creating average salary column
ds_jobs_uncleaned['avg_salary']=(ds_jobs_uncleaned.min_salary + ds_jobs_uncleaned.max_salary)//2
```

```python
ds_jobs_uncleaned.head()
```

```python
# cleaning company column
ds_jobs_uncleaned['Company Name'] = ds_jobs_uncleaned['Company Name'].apply(lambda x : x.split('\n')[0])
ds_jobs_uncleaned.head()
```

```python
ds_jobs_uncleaned.sort_values(by='Rating', ascending = True)
```

```python
# importing numpy library
import numpy as np
```

```python
# change the rating values whenever they equal -1.0
ds_jobs_uncleaned.Rating = np.where(ds_jobs_uncleaned.Rating == -1.0, 0, ds_jobs_uncleaned.Rating)
```

```python
ds_jobs_uncleaned.sort_values(by='Rating', ascending = True)
```

```python
# separate state from location and create its own column
ds_jobs_uncleaned['state'] = ds_jobs_uncleaned['Location'].apply(lambda x : x.split(',')[-1])
```

```python
ds_jobs_uncleaned.head()
```

```python
# create new column same_state anf if it's the same as hq, put 1 else 0
ds_jobs_uncleaned['same_state'] = ds_jobs_uncleaned.apply(lambda x : 1 if x.Location == x.Headquarters else 0, axis=1)
```
